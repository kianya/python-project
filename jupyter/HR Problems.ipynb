## T-test статистика
С использованием t-test статистики проверить взаимосвязи между несколькими переменными

```python
## Проверка гипотезы: уровень удовлетворенности сотрудников, участвующих в одном проекте не зависят от департамента
# Для примера возьмем проект 2 и остальные проекты для департамента hr

hr_prog2 = hr[(hr["number_project"]==2)]
hr_all_prog_without2 = hr[(hr["number_project"]!=2)]

sc.stats.ttest_ind(hr_prog2['satisfaction_level'], hr_all_prog_without2['satisfaction_level']) 
```

Согласно полученному значению pvalue, уровень удовлетворенности в проекте 2 сильно отличается от удовлетворенности участников департамента hr на других проектах.

Проверим ту же гипотезу для департамента IT

```python
IT = df[(df["department"]=='IT')]

IT_prog2 = IT[(IT["number_project"]==2)]
IT_all_prog_without2 = IT[(IT["number_project"]!=2)]


sc.stats.ttest_ind(IT_prog2['satisfaction_level'], IT_all_prog_without2['satisfaction_level']) 
```

Для департамента IT, так же удовлетворенность в проекте 2 сильно отличается от удовлетворенности другими проектами.

```python
## Проверка гипотезы: уровень удовлетворенности на проекте 2 в департаменте IT и HR одинаковы

sc.stats.ttest_ind(IT_prog2['satisfaction_level'], hr_prog2['satisfaction_level']) 
```

#### Гипотеза 2

    В департаменте продаж (sales) люди уходят из-за неудовлетворенности в зарплате.

```python
# Корреляция для департамента продаж
sales = df[(df['department']=='sales') & (df['number_project']!=2) & (df['number_project']!=6)]
corr - sales.corr()

sns.heatmap(corr, annot=True,
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values,
            linewidths=.5)
```

Исходя из матрица корреляции видно, что в департаменте продаж (для сотрудников, не привлеченных на проект 2), уход сотрудников связан с уровнем неудовлетворенности, отсутсвием повышения. В то же время чем больше сотрудник работает, тем скорее он добивается повышения, и тем больше он удовлетворен работой.

При этом корреляция между зарплатой и уходом сотрудника из кампании не значительна.

```python
## Проверка гипотезы: уход из кампании не зависит от уровня зарплаты
sales_low = sales[(sales['salary']=='low')]
sales_mh = sales[(sales['salary']=='medium') | (sales['salary']=='high')]

sc.stats.ttest_ind(sales_low['left'], sales_mh['left'])
```
